```python
# data_dict_csv = pd.read_csv(r"2025_Problem_C_Data\data_dictionary.csv")
athletes_csv = pd.read_csv(r"2025_Problem_C_Data\summerOly_athletes.csv")
hosts_csv = pd.read_csv(r"2025_Problem_C_Data\summerOly_hosts.csv")
medal_counts_csv = pd.read_csv(r"2025_Problem_C_Data\summerOly_medal_counts.csv")
programs_csv = pd.read_csv(r"2025_Problem_C_Data\summerOly_programs.csv", encoding='windows-1252')
```

```python
# Print heads of each file
print("Athletes CSV Head:\n", athletes_csv.head())
print("Hosts CSV Head:\n", hosts_csv.head())
print("Medal Counts CSV Head:\n", medal_counts_csv.head())
print("Programs CSV Head:\n", programs_csv.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming medal_counts_csv has columns 'Year' and 'Total'
sns.barplot(data=medal_counts_csv, x="Year", y="Total")
plt.title("Total Medals by Year")
plt.xticks(rotation=45)
plt.show()

```

```python
# Assuming medal_counts_csv has columns 'NOC', 'Gold'
most_gold_country = medal_counts_csv.groupby("NOC")["Gold"].sum().idxmax()
print(f"The country with the most gold medals is: {most_gold_country}")
```
